## Research Agent applied to generate a RAG Database

#### Importing libs

```python
from google import genai
from google.genai import types
from IPython.display import display, Markdown, HTML
import json

import os
from pathlib import Path
import sys
sys.path.append(str(Path().resolve().parent.parent))
from utils.config import CLIENT, MODEL
from agents.research_agent import research_agent
from utils.load_prompt import load_prompt
from utils.format import parse_json_output
```

#### Generating the database

```python
neighborhoods = ['Vila Mariana', 'Mooca', 'Brooklin', 'Tatuapé', 'Santana']
results = []

for neighborhood in neighborhoods:
    result = research_agent(neighborhood)["result"]

    prompt = load_prompt("chunk_agent_v1").replace("{relatorio}", result)
    response = CLIENT.models.generate_content(
            model=MODEL,
            contents=prompt,
            config=types.GenerateContentConfig(temperature=0.2),
        )

    results.extend(parse_json_output(response.text))
```

```python
out_path = Path("/Users/vinicius.araujo/Github/genai/data") / "rag_database.json"
out_path.write_text(json.dumps(results, ensure_ascii=False, indent=2), encoding="utf-8")
```
