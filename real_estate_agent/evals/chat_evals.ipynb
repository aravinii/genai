## Chat evals

#### Importing libs

```python
from google import genai
from google.genai import types

import os
import json
from pathlib import Path
import sys
import re
sys.path.append(str(Path().resolve().parent))

from utils.load_prompt import load_prompt
from utils.format import format_judge_conversation, parse_json_output
from utils.config import CLIENT, MODEL, CONVERSATION_PATH, CONVERSATION_PATH_V2
```

#### Applying LLM as a judge evaluation framework

```python
#conversations = json.loads(CONVERSATION_PATH.read_text(encoding="utf-8"))
conversations = json.loads(CONVERSATION_PATH_V2.read_text(encoding="utf-8"))
```

```python
#evals_results = []
i = 0
for conversation in conversations:
    print(i)
    #if i >= 3:
    conversation_history = format_judge_conversation(conversation)
    prompt = load_prompt("quality_agent_v1").replace("{conversation_history}", conversation_history)

    response = CLIENT.models.generate_content(
            model=MODEL,
            contents=prompt,
            config=types.GenerateContentConfig(temperature=0.2),
        )
    score = parse_json_output(response.text)
    score["approach"] = 'RAG Agent'
    evals_results.append(score)
    i += 1
```

```python
out_path = Path("/Users/vinicius.araujo/Github/genai/data") / "chat_evaluations.json"
out_path.write_text(json.dumps(evals_results, ensure_ascii=False, indent=2), encoding="utf-8")
```
