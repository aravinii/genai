## LLM as a judge

#### Importing libs

```python
from google import genai
from google.genai import types

import os
import json
from pathlib import Path
import sys
import re
sys.path.append(str(Path().resolve().parent))

from utils.load_prompt import load_prompt
from utils.format import format_judge_conversation, parse_json_output
from utils.config import CLIENT, MODEL, CONVERSATION_PATH
```

#### Defining auxiliary functions

```python
conversations = json.loads(CONVERSATION_PATH.read_text(encoding="utf-8"))
```

```python
evals_results = []
conversation = conversations[1]
conversation_history = format_judge_conversation(conversation)
prompt = load_prompt("quality_agent_v1").replace("{conversation_history}", conversation_history)

response = CLIENT.models.generate_content(
        model=MODEL,
        contents=prompt,
        config=types.GenerateContentConfig(temperature=0.2),
    )
raw = response.text
evals_results.append(parse_json_output(raw))
```
